```python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼"
    except TypeError:
        return "éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å­—ï¼"
    finally:
        # ç„¡è«–å¦‚ä½•éƒ½æœƒåŸ·è¡Œ
        print("æª”æ¡ˆæ“ä½œå®Œæˆ")
        # é€™è£¡å¯ä»¥æ”¾æ¸…ç†ç¨‹å¼ç¢¼ï¼Œå¦‚é—œé–‰æª”æ¡ˆ

print(safe_divide(10, 2))    # 5.0
print(safe_divide(10, 0))    # éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼
print(safe_divide(10, "a"))  # éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å­—ï¼
```

```python
def to_integer(value):
    try:
        return int(value)
    except ValueError:
        return "éŒ¯èª¤ï¼šè¼¸å…¥çš„ä¸æ˜¯æ•´æ•¸ï¼"
    finally:
        print("è™•ç†å®Œæˆ")
        #print("è™•ç†å®Œæˆï¼š{value}")
        print(f"è™•ç†å®Œæˆï¼š{value}")
print(to_integer("100"))   # 100
print(to_integer("5.5"))
print(to_integer("abc"))   # éŒ¯èª¤ï¼šè¼¸å…¥çš„ä¸æ˜¯æ•´æ•¸ï¼
```

```python
def read_file(filename):
    try:
        f = open(filename, 'r', encoding='utf-8')
        content = f.read()
        return content
    except FileNotFoundError:
        return "éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°æª”æ¡ˆï¼"
    except UnicodeDecodeError:
        return "éŒ¯èª¤ï¼šæª”æ¡ˆç·¨ç¢¼ä¸æ­£ç¢ºï¼"
    finally:
        print("å·²å˜—è©¦é–‹å•Ÿæª”æ¡ˆ")
print(read_file("data.txt"))
print(read_file("ä¸å­˜åœ¨çš„æª”æ¡ˆ.txt"))
```
